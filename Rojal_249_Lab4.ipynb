```python
class domain:
    def __init__(self,domainname):
        self.dom=domainname
       
class platform:
    def __init__(self,platforms):
        self.plat=platforms


        
#Multiple Inheritance

class game(domain,platform):
    def __init__(self,gamename1):
        domain.__init__(self,domainname)
        platform.__init__(self,platforms)
        self.gamename=gamename1   
    def display(self):
        print(self.dom,"\n",self.plat)

#MultiLevel Inheritance

class fram(game):
    def __init__(self):
        domain.__init__(self,domainname)
        platform.__init__(self,platforms)
        game.__init__(self,gamename1)
    def show(self):
        print(self.dom,"\n",self.plat,"\n",self.gamename)



#Hierarchical Inheritance

class game1(platform):
    def __init__(self,gamename2):
        platform.__init__(self,platforms)
        self.gname1=gamename2
    def view1(self):
        print("Game1 : ",self.gname1,"\n Platform : ",self.plat)
        
class game2(platform):
    def __init__(self,gamename3):
        platform.__init__(self,platforms)
        self.gname2=gamename3
    def view2(self):
        print("Game2 : ",self.gname2,"\n Platform : ",self.plat)



#input
domainname=input("Enter your domain : ")
platforms=input("Enter the platform you selected : ")
gamename1=input("Enter the game name : ")


a=domain(domainname)
b=platform(platforms)
print("\nThis is the result of multiple inhertance : ")
multiple=game(gamename1)
multiple.display()

print("\nThis is the result of multilevel inhertance : ")
level=fram()
level.show()

print("\nThis is the result of Hierarchical inhertance : ")
gamename2=input("Enter the first game name : ")
gamename3=input("Enter the second game name : ")

hierachy1=game1(gamename2)
hierachy1.view1()

hierachy2=game2(gamename3)
hierachy2.view2()
```
